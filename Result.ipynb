# KẾT QUẢ

```python
import pandas as pd

result_svm_df = pd.read_csv("SVM/result-svm.csv")
result_knn_df = pd.read_csv("KNN/result-knn.csv")
result_rf_df = pd.read_csv("RF/result-rf.csv")
result_xgb_df = pd.read_csv("XGB/result-xgb.csv")
result_lr_df = pd.read_csv("LR/result-lr.csv")
result_dt_df = pd.read_csv("DT/result-dtc.csv")
result_ada_df = pd.read_csv("AB/result-ada.csv")
```

### K áp dụng PCA

```python
a = pd.concat([result_svm_df, result_knn_df, result_lr_df], axis = 1).drop(columns = ["Unnamed: 0", "support"]).round(3)
groups = ['SVM']*3 + ['KNN']*3  + ['LogisticRegression']*3 
metrics = ['precision', 'recall', 'f1-score'] * 3

result = pd.MultiIndex.from_arrays([groups, metrics], names=['Group', 'Metric'])
a.columns = result
a.index = result_svm_df["Unnamed: 0"].values
a = a.drop('accuracy', axis = 0)
a
```

```python
b = pd.concat([result_rf_df, result_dt_df, result_ada_df, result_xgb_df], axis = 1).drop(columns = ["Unnamed: 0", "support"]).round(3)
groups = ['RandomForest']*3 + ['DecisionTree']*3 + ['AdaBoost']*3  + ['XGBoost']*3
metrics = ['precision', 'recall', 'f1-score'] * 4

result = pd.MultiIndex.from_arrays([groups, metrics], names=['Group', 'Metric'])
b.columns = result
b.index = result_svm_df["Unnamed: 0"].values
b = b.drop('accuracy', axis = 0)
b
```
