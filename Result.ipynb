#### Với bộ dữ liệu khi qua lớp full connected (gồm 6400 features) và áp dụng PCA (n = 3967) => 99% thông tin

SVM

```python
result_svm_pca_99_df = pd.read_csv("./SVM/result-svm-pca-99.csv")
result_svm_pca_99_df
```

KNN

```python
result_knn_pca_2_df = pd.read_csv("./KNN/result-knn-pca-99.csv")
result_knn_pca_2_df
```

```python
a = pd.concat([result_svm_df, result_knn_df, result_knn_pca_2_df, result_knn_knn_2_df, result_svm_pca_99_df, result_knn_pca_2_df], axis = 1).drop(columns = ["Unnamed: 0", "support"]).round(3)
groups = ['SVM']*3 + ['KNN']*3 + ['SVM_PCA_2']*3 + ['KNN_PCA_2']*3 + ['SVM_PCA_99']*3 + ['KNN_PCA_99']*3
metrics = ['precision', 'recall', 'f1-score'] * 6

result = pd.MultiIndex.from_arrays([groups, metrics], names=['Group', 'Metric'])
a.columns = result
a.index = result_svm_df["Unnamed: 0"].values
a
```

SVM, KNN: Kết quả khi train với data là data đi qua lớp FC<br>
SVM_PCA_2, KNN_PCA_2:  Kết quả khi train với data là data đi qua lớp FC sau đó áp dụng PCA (n = 2)<br>
SVM_PCA_99, KNN_PCA_99:  Kết quả khi train với data là data đi qua lớp FC sau đó áp dụng PCA để giữ lại 99% thông tin
